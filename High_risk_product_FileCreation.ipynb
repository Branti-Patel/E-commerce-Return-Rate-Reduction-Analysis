```python
import pandas as pd
```

```python
products = pd.read_csv("C:\\Users\\HP\\Downloads\\Products.csv")
returns = pd.read_csv("C:\\Users\\HP\\Downloads\\Returns_Refund.csv")
rating = pd.read_csv("C:\\Users\\HP\\Downloads\\Ratings.csv")
```

```python
total_orders_per_product = returns.groupby('Or_ID').size().rename('total_returns')
all_orders = pd.read_csv("C:\\Users\\HP\\Downloads\\Orders.csv")
total_orders = all_orders.groupby('Or_ID').size().rename('total_orders')
```

```python
return_rate = pd.concat([total_orders, total_orders_per_product], axis=1).fillna(0)
return_rate['return_rate'] = return_rate['total_returns'] / return_rate['total_orders']
```

```python
print(return_rate)
```

```python
ratings = pd.read_csv("C:\\Users\\HP\\Downloads\\Ratings.csv")
```

```python
avg_ratings = ratings.groupby('Or_ID')['Prod_Rating'].mean().rename('avg_rating')
```

```python
risk_df = pd.concat([return_rate, avg_ratings], axis=1).fillna(0)
```

```python
high_risk = risk_df[(risk_df['return_rate'] >= 0.2) | (risk_df['avg_rating'] < 3.0)]
```

```python
high_risk = high_risk.merge(products[['Or_ID', 'Category']], on='Category', how='left')
```

```python
high_risk.to_csv("High_Risk_Products.csv", index=False)
```
